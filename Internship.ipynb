```python
import PyPDF2

def remove_url_from_pdf(input_path, output_path, url_to_remove):
    with open(input_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        writer = PyPDF2.PdfWriter()

        for page in reader.pages:
            annotations = page.get('/Annots')
            if annotations:
                if not isinstance(annotations, list):
                    annotations = [annotations]
                new_annotations = []
                for annotation in annotations:
                    if isinstance(annotation, PyPDF2.generic.IndirectObject):
                        annotation = annotation.get_object()
                    if isinstance(annotation, dict):
                        if '/A' in annotation:
                            uri = annotation['/A']
                            if '/URI' in uri and uri['/URI'].startswith(url_to_remove):
                                continue
                    new_annotations.append(annotation)
                page.Annots = new_annotations
            writer.add_page(page)

        with open(output_path, 'wb') as output_file:
            writer.write(output_file)

# Example usage
pdf_input_path = '/content/drive/MyDrive/Major_1/Book.pdf'
pdf_output_path = '/content/drive/MyDrive/Major_1/Book_1.pdf'
url_to_remove = "http://freebooks.by.ru/view/CProgrammingLanguage/kandr.html (1 of 5) [9/6/2002 12:20:42 ]"

remove_url_from_pdf(pdf_input_path, pdf_output_path, url_to_remove)
```

```python
import re
import fitz  # PyMuPDF

def delete_c_code_and_save(file_path, output_file_path):
    # Open the PDF file
    pdf_document = fitz.open(file_path)

    # Create a new PDF document to store modified pages
    modified_pdf = fitz.open()

    # Iterate through pages
    for page_number in range(pdf_document.page_count):
        # Get the page
        page = pdf_document[page_number]

        # Extract text from the page
        page_text = page.get_text()

        # Define the regular expression pattern to match C code
        c_code_pattern = (
            r'#include\s*<stdio.h>\s*.*?(?:int\s+)?main\s*\(\s*\)[^{]*\{(?:[^}]*\}|[^}]*\n*)\}'
            r'|(?:int|float|char|void|double|short|long|if|else|for|while|do|switch|case|break|continue|return|'
            r'size_t|struct|typedef|enum|volatile|static|const|auto|register)\s+\w+\s*\([^)]*\)[^}]*\{(?:[^}]*\}|[^}]*\n*)\}'
            r'|\b(?:printf|puts)\s*\([^;]*\)\s*;|if\s*\([^)]*\)[^{]*\{(?:[^}]*\}|[^}]*\n*)\}'
            r'|\belse\s*[^{]*\{(?:[^}]*\}|[^}]*\n*)\}'
            r'|\b(?:for|while|do)\s*\([^;]*\)[^{]*\{(?:[^}]*\}|[^}]*\n*)\}'
        )

        # Delete the matched C code from the page text
        modified_text = re.sub(c_code_pattern, '', page_text, flags=re.MULTILINE | re.DOTALL)

        # Create a new page in the modified PDF with the modified text
        modified_page = modified_pdf.new_page(width=page.rect.width, height=page.rect.height)
        modified_page.insert_text((10, 10), modified_text)

    # Save the modified PDF to the output file
    modified_pdf.save(output_file_path)
    modified_pdf.close()

    print("C code deleted from the PDF and saved to:", output_file_path)

# Example usage
input_file_path = '/content/drive/MyDrive/Major_1/Book.pdf'
output_file_path = '/content/drive/MyDrive/Major_1/Book_1.pdf'
delete_c_code_and_save(input_file_path, output_file_path)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def calculate_cosine_similarity(text1, text2):
    vectorizer = CountVectorizer().fit_transform([text1, text2])
    vectors = vectorizer.toarray()
    similarity = cosine_similarity(vectors)
    return similarity[0, 1]

def remove_non_printable(text):
    # Remove non-printable characters and special characters
    printable = set(string.printable)
    adjusted_text = ''.join(filter(lambda x: x in printable, text))

    return adjusted_text

def remove_urls_and_special_chars(text):
    # Remove URLs starting with "http://" or "https://"
    adjusted_text = re.sub(r"http[s]?://\S+", "", text)

    # Remove links in the format: "http://freebooks.by.ru/view/CProgrammingLanguage/chapter1.html (13 of 30) [9/6/2002 12:21:11 ]"
    adjusted_text = re.sub(r"http\S+\s+\(\d+\s+of\s+\d+\)\s+\[\d{1,2}/\d{1,2}/\d{4}\s+\d{1,2}:\d{2}:\d{2}\s+\]", "", adjusted_text)

    # Remove non-printable characters
    adjusted_text = remove_non_printable(adjusted_text)

    return adjusted_text

def remove_specific_url_from_summary(summary):
    # Remove the specific URL from the summary
    specific_url_pattern = r"http[s]?://freebooks\.by\.ru/view/CProgrammingLanguage/chapter1\.html"
    summary_without_url = re.sub(specific_url_pattern, "", summary)

    return summary_without_url

def summarize(text, per):
    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)
    tokens = [token.text for token in doc]
    word_frequencies = {}
    for word in doc:
        if word.text.lower() not in list(STOP_WORDS):
            if word.text.lower() not in punctuation:
                if word.text not in word_frequencies.keys():
                    word_frequencies[word.text] = 1
                else:
                    word_frequencies[word.text] += 1
    max_frequency = max(word_frequencies.values()) if word_frequencies else 1
    for word in word_frequencies.keys():
        word_frequencies[word] = word_frequencies[word] / max_frequency
    sentence_tokens = [sent for sent in doc.sents]
    sentence_scores = {}
    for sent in sentence_tokens:
        for word in sent:
            if word.text.lower() in word_frequencies.keys():
                if sent not in sentence_scores.keys():
                    sentence_scores[sent] = word_frequencies[word.text.lower()]
                else:
                    sentence_scores[sent] += word_frequencies[word.text.lower()]
    select_length = int(len(sentence_tokens) * per)
    summary = nlargest(select_length, sentence_scores, key=sentence_scores.get)
    final_summary = [word.text for word in summary]
    summary = ' '.join(final_summary)
    return summary

def generate_summary_from_book(book_path, start_page, end_page, topics, summary_percentage):
    # Extract text from PDF
    pdf_reader = PyPDF2.PdfReader(book_path)
    extracted_text = ""
    for i in range(start_page - 1, end_page):
        extracted_text += pdf_reader.pages[i].extract_text()

    # Remove URLs, specific strings, and special characters from extracted text
    extracted_text_cleaned = remove_urls_and_special_chars(extracted_text)

    # Create a dictionary to store topic-summary pairs
    topic_summary_dict = {}

    # Generate summary for each topic
    for i, topic in enumerate(topics):
        # Find the start and end positions of the current topic
        start_pos = extracted_text_cleaned.find(topic)
        end_pos = extracted_text_cleaned.find(topics[i + 1]) if i < len(topics) - 1 else len(extracted_text_cleaned)

        # Extract the relevant text for the current topic
        topic_text = extracted_text_cleaned[start_pos:end_pos]

        # Check if there is any text available for the current topic
        if topic_text:
            # Generate the summary for the current topic
            summary = summarize(topic_text, summary_percentage)

            # Remove the specific URL from the summary
            summary = remove_specific_url_from_summary(summary)

            # Store the topic-summary pair in the dictionary
            topic_summary_dict[topic] = summary

    # Return the topic-summary dictionary
    return topic_summary_dict

for i, topic in enumerate(topics):
    start_pos = extracted_text_cleaned.find(topic)
    end_pos = extracted_text_cleaned.find(topics[i + 1]) if i < len(topics) - 1 else len(extracted_text_cleaned)

    topic_text = extracted_text_cleaned[start_pos:end_pos]

# Provide the necessary inputs
book_path = '/content/drive/MyDrive/Major_1/Book_1.pdf'
start_page = 13
end_page = 43
topics = [
    "Getting Started",
    "Variables and Arithmetic Expressions",
    "The for statement",
    "Symbolic Constants",
    "Character Input and Output",
    "File Copying",
    "Character Counting",
    "Line Counting",
    "Word Counting",
    "Arrays",
    "Functions",
    "Arguments - Call by Value",
    "Character Arrays",
    "External Variables and Scope"
]
summary_percentage = 0.3

# Generate the summary from the book
summary_dict = generate_summary_from_book(book_path, start_page, end_page, topics, summary_percentage)

# ... (previous code)

# Display the topic and summary for each topic along with cosine similarity
for topic, summary in summary_dict.items():
    print(" --------------------------------------------  ")
    print(f"{topic}")
    print(" --------------------------------------------  ")
    print(f" {summary}")
    print(" --------------------------------------------  ")

    # Calculate cosine similarity between topic_text and summary
    similarity_score = calculate_cosine_similarity(topic_text, summary)
    print(f"Cosine Similarity: {similarity_score}")

    print(" --------------------------------------------  ")
    print(" --------------------------------------------  ")
```
